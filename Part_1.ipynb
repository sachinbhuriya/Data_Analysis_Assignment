### Task 1 - Develop a function to read a single column from a CSV file

````python
import csv

def read_single_column(file_path, column_name):
    """Reads a single column from a CSV file."""
    column_data = []
    try:
        with open(file_path, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                column_data.append(float(row[column_name]))
    except KeyError:
        print(f"Column '{column_name}' not found.")
    return column_data

# Example usage:
data = read_single_column("/content/part_1.csv", "poverty_rate")
print(data)
````

### Task 2 - Develop a function to read CSV data from a file into memory

````python
def read_csv_to_memory(file_path):
    """Reads an entire CSV file into memory as a list of dictionaries."""
    data = []
    try:
        with open(file_path, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                # Convert numeric fields to float where applicable
                data.append({key: float(value) if value.replace('.', '', 1).isdigit() else value for key, value in row.items()})
    except FileNotFoundError:
        print(f"File '{file_path}' not found.")
    return data

# Example usage:
dataset = read_csv_to_memory("part_1.csv")
print(dataset)
````

### Task 3 - Develop a function to calculate the Pearson Correlation Coefficient for two lists of data

````python
from scipy.stats import pearsonr

def calculate_pearson_correlation(list1, list2):
    """Calculates Pearson Correlation Coefficient between two lists."""
    if len(list1) != len(list2):
        raise ValueError("Lists must have the same length.")
    correlation, _ = pearsonr(list1, list2)
    return correlation

# Example usage:
corr = calculate_pearson_correlation([14.1, 28.8, 14.6], [0.097, 0.065, 0.166])
print(f"Pearson Correlation: {corr}")
````
